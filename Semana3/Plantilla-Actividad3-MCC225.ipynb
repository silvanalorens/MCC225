#### **6. Mini simulador de decisión**

La siguiente función parte de la síntesis de los Cuadernos 3 al 8 y convierte esa lectura conceptual del bloque en una regla sencilla de recomendación.

**Tu tarea:** ajustar pesos, criterios y mensajes hasta que la herramienta tenga sentido para tu problema.


```python
familias = pd.DataFrame([
    {"Familia": "Late Fusion", "clasificacion": 3, "retrieval": 0, "temporal": 0, "no_alineado": 0, "bajo_computo": 5, "interaccion_fina": 1},
    {"Familia": "MMBT-Lite", "clasificacion": 4, "retrieval": 1, "temporal": 0, "no_alineado": 1, "bajo_computo": 2, "interaccion_fina": 3},
    {"Familia": "Deep Fusion Transformer", "clasificacion": 5, "retrieval": 1, "temporal": 1, "no_alineado": 1, "bajo_computo": 1, "interaccion_fina": 5},
    {"Familia": "Bi-encoder contrastivo", "clasificacion": 1, "retrieval": 5, "temporal": 1, "no_alineado": 1, "bajo_computo": 4, "interaccion_fina": 2},
    {"Familia": "VideoCLIP-lite / ALPRO-lite", "clasificacion": 1, "retrieval": 4, "temporal": 5, "no_alineado": 2, "bajo_computo": 1, "interaccion_fina": 4},
    {"Familia": "MulT-lite / SPT-lite", "clasificacion": 3, "retrieval": 1, "temporal": 4, "no_alineado": 5, "bajo_computo": 1, "interaccion_fina": 5},
])

def arquitectura_recomendada(
    task="clasificacion",
    temporal=False,
    no_alineado=False,
    bajo_computo=True,
    interaccion_fina=False,
    top_k=3
):
    weights = {
        "clasificacion": 3 if task == "clasificacion" else 0,
        "retrieval": 3 if task == "retrieval" else 0,
        "temporal": 2 if temporal else 0,
        "no_alineado": 2 if no_alineado else 0,
        "bajo_computo": 2 if bajo_computo else 0,
        "interaccion_fina": 2 if interaccion_fina else 0,
    }

    scored = familias.copy()
    scored["score"] = 0.0
    for col, w in weights.items():
        scored["score"] += scored[col] * w

    return scored.sort_values("score", ascending=False).head(top_k)[["Familia", "score"]]

arquitectura_recomendada(task="clasificacion", temporal=False, no_alineado=False, bajo_computo=True, interaccion_fina=False)
```

```python
# Prueba distintos escenarios y comenta si el resultado te parece razonable.
escenarios = [
    {"task": "clasificacion", "temporal": False, "no_alineado": False, "bajo_computo": True, "interaccion_fina": False},
    {"task": "retrieval", "temporal": False, "no_alineado": False, "bajo_computo": True, "interaccion_fina": False},
    {"task": "retrieval", "temporal": True, "no_alineado": False, "bajo_computo": False, "interaccion_fina": True},
    {"task": "clasificacion", "temporal": True, "no_alineado": True, "bajo_computo": False, "interaccion_fina": True},
]

for i, sc in enumerate(escenarios, 1):
    print(f"Escenario {i}: {sc}")
    display(arquitectura_recomendada(**sc))
    print("-" * 60)
```

```python
# Opcional: versión interactiva con ipywidgets
try:
    import ipywidgets as widgets
    from IPython.display import display

    def _interactive(task, temporal, no_alineado, bajo_computo, interaccion_fina, top_k):
        display(arquitectura_recomendada(
            task=task,
            temporal=temporal,
            no_alineado=no_alineado,
            bajo_computo=bajo_computo,
            interaccion_fina=interaccion_fina,
            top_k=top_k
        ))

    widgets.interact(
        _interactive,
        task=widgets.Dropdown(options=["clasificacion", "retrieval"], value="clasificacion"),
        temporal=False,
        no_alineado=False,
        bajo_computo=True,
        interaccion_fina=False,
        top_k=widgets.IntSlider(value=3, min=1, max=6, step=1)
    )
except Exception as e:
    print("ipywidgets no está disponible en este entorno:", e)
```

#### **7. Visualización comparativa**

Puedes usar este gráfico como punto de partida para pensar la relación entre **costo** e **interacción fina**.  
Modifica los valores si no te convencen.


```python
mapa_costo = pd.DataFrame([
    {"Modelo": "Late Fusion", "Interacción fina": 1, "Costo": 1},
    {"Modelo": "MMBT-Lite", "Interacción fina": 3, "Costo": 3},
    {"Modelo": "Deep Fusion", "Interacción fina": 5, "Costo": 4},
    {"Modelo": "Bi-encoder", "Interacción fina": 2, "Costo": 2},
    {"Modelo": "VideoCLIP/ALPRO-lite", "Interacción fina": 4, "Costo": 4},
    {"Modelo": "MulT/SPT-lite", "Interacción fina": 5, "Costo": 5},
])

fig, ax = plt.subplots()
ax.scatter(mapa_costo["Costo"], mapa_costo["Interacción fina"], s=90)

for _, row in mapa_costo.iterrows():
    ax.text(row["Costo"] + 0.03, row["Interacción fina"] + 0.03, row["Modelo"], fontsize=9)

ax.set_xlabel("Costo relativo")
ax.set_ylabel("Interacción fina")
ax.set_title("Mapa didáctico: costo vs. interacción")
ax.grid(True, alpha=0.3)
plt.show()
```
